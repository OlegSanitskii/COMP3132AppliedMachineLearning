###10.	Import models and layers from the keras library.

````python
from tensorflow.keras import models, layers
````

###11.	Define a sequential model and call it myNetwork.

````python
myNetwork = models.Sequential()
````

###12.	Reshape the images from 28x28 to one column with 784 neurons (flattening)

````python
myNetwork.add(layers.Flatten(input_shape=(28, 28)))
````

###13.	Normalize the image by dividing the image by 255

````python
train_images = train_images / 255.0
test_images = test_images / 255.0
````

###14.	Add one hidden layer that has 512 neurons, using the relu activation function.

````python
myNetwork.add(layers.Dense(512, activation='relu'))
````

###15.	Add another hidden layer that has 128 neurons, using the relu activation function.

````python
myNetwork.add(layers.Dense(128, activation='relu'))
````

###16.	Add the last layer as a 10-neuron dense layer that uses the softmax as the activation function. Why do we use softmax for the last layer? How does it work under the hood? Make sure to explain this in your video.

```text
Softmax is used in the last layer because this is a multi-class classification
problem with 10 possible classes. The softmax function converts the raw output
values of the network into probabilities for each class.

Each output neuron represents one class, and softmax ensures that all output
values are between 0 and 1 and that their sum is equal to 1. This makes it easy
to interpret the models output as probabilities and to select the class with
the highest probability as the final prediction.

````python
myNetwork.add(layers.Dense(10, activation='softmax'))

myNetwork.summary()
````

###17.	Use the following two settings for the compiler and run them separately and see what the differences are.

**Option A**
```text
i.   Optimizer: adam
ii.  loss: 'sparse_categorical_crossentropy'
iii. metrics: ['accuracy']
```

**Option B**
```text
i.   Optimizer: rmsprop
ii.  loss: 'categorical_crossentropy'
iii. metrics: ['accuracy']
```

````python
# Option A

def build_model():
    model = models.Sequential([
        layers.Flatten(input_shape=(28, 28)),
        layers.Dense(512, activation='relu'),
        layers.Dense(128, activation='relu'),
        layers.Dense(10, activation='softmax')
    ])
    return model

model_A = build_model()
model_A.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
````

````python
# Option B


from tensorflow.keras.utils import to_categorical

train_labels_cat = to_categorical(train_labels, num_classes=10)
test_labels_cat = to_categorical(test_labels, num_classes=10)

model_B = build_model()
model_B.compile(
    optimizer='rmsprop',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
````

### 18.	Now after the compilation, please try to find the pattern using the fit command. Set the number of epochs to 10.

````python
# Option A
history_A = model_A.fit(train_images, train_labels, epochs=10, validation_split=0.2)
````

````python
# Option B
history_B = model_B.fit(train_images, train_labels_cat, epochs=10, validation_split=0.2)
````

### 19.	How do you compare the fashion_MNIST with what we learned in the class using the MNIST? Explain in your video.


```text
Fashion MNIST is more difficult than the original MNIST dataset.
In MNIST, handwritten digits usually have very clear and distinct shapes.
In Fashion MNIST, many clothing items look similar to each other.

For example, shirts, t-shirts, pullovers, and coats often have overlapping
shapes and outlines, which makes it harder for the model to distinguish
between classes. Because of this, Fashion MNIST typically results in lower
accuracy compared to MNIST.

### 20.	What can we infer from the differences in the accuracy? What could be the reasons for that? Explain in your video.

```text
The lower accuracy on Fashion MNIST happens because the classification task
is more complex. Many classes have similar shapes, so the model can easily
confuse them.

In addition, this model uses only fully connected layers and does not take
advantage of spatial information in the images. Using convolutional neural
networks would likely improve the accuracy. Finally, Fashion MNIST images
show more variation in shape and style than MNIST digits, which increases
the difficulty of the problem.

### 21.	Use the evaluate() function from the Keras library to calculate the achieved accuracy and loss over the test images and labels. Do we have overfitting? Explain in your video.

````python
# Option A

test_loss_A, test_acc_A = model_A.evaluate(test_images, test_labels)
print("Option A - Test loss:", test_loss_A)
print("Option A - Test accuracy:", test_acc_A)
````

````python
# Option B

test_loss_B, test_acc_B = model_B.evaluate(test_images, test_labels_cat)
print("Option B - Test loss:", test_loss_B)
print("Option B - Test accuracy:", test_acc_B)
````
